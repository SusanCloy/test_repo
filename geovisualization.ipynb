```python
gapminder = px.data.gapminder()
display(gapminder.head(), gapminder.shape)
```

```python
gapminder.info()
```

Question 1: Map out the mean life expectancy of all the Asian countries according to the data.

```python
asia_life = gapminder.query('continent == "Asia"').groupby('country')['lifeExp'].mean().reset_index()
asia_life.head(3)
```

```python
fig = px.choropleth(asia_life, locations=asia_life['country'].values.tolist(), locationmode="country names", scope="asia", 
                    color='lifeExp', hover_name='country', color_continuous_scale = px.colors.sequential.Plasma, 
                    title='Average Life Expectancy of Asian Countries',) 

fig.show(renderer='colab')
```

# Question 3: Map out deviation in GDP of each country in Europe and South America.

```python
gapminder.columns
```

```python
data = gapminder.query('continent == ["Europe", "Americas"]').groupby('country')['gdpPercap'].std().reset_index().rename(columns= {'gdpPercap': 'Gdp_Deviation'})
data.head()
```

```python
fig = px.scatter_geo(data, color='Gdp_Deviation', locations=data['country'].values.tolist(), locationmode='country names',
                     hover_name='country', size='Gdp_Deviation', scope='world', title='Deviation of GDP of Europe And American Countries', )

fig.show(renderer='colab')
```

# Question 3: Map out the change in population of each African country in the last 3 decades.

```python
africa = gapminder.query('continent == "Africa" and year >= 1977')
africa.head()
```

```python
len(africa['country'].unique())
```

```python
fig = px.scatter_geo(africa, locations=africa['country'].values.tolist(), locationmode='country names', animation_frame='year', 
            scope='africa', title='Change in Population of each African Country Over the Last 3 Decades', color='country', 
            hover_name='country', color_continuous_scale = px.colors.sequential.Plasma, size=africa['pop'] * 100)

fig.show(renderer='colab')
```

```python
fig = px.choropleth(africa, locations=africa['country'].values.tolist(), locationmode='country names', animation_frame='year', 
            scope='africa', title='Change in Population of each African Country Over the Last 3 Decades', color='pop',  
            hover_name='country', color_continuous_scale = px.colors.sequential.Plasma,)

fig.show(renderer='colab')
```
